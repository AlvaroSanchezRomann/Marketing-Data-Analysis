### Crea y evalúa los modelos

````python
#Escribe tu código

import pandas as pd
from google.colab import files
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Leer el archivo CSV
df = pd.read_csv('creditcard.csv')

# Separar las características (X) y la etiqueta (y)
X = df.drop('Class', axis=1)
y = df['Class']

# Dividir el dataset en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear el modelo de Regresión Logística
logistic_model = LogisticRegression(max_iter=1000)  # Incrementar max_iter si es necesario

# Entrenar el modelo
logistic_model.fit(X_train, y_train)

# Predecir con el modelo
y_pred_logistic = logistic_model.predict(X_test)

# Evaluar el modelo
print("Evaluación del Modelo de Regresión Logística")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_logistic))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_logistic))
print(f"Accuracy: {accuracy_score(y_test, y_pred_logistic):.2f}")

````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_recall_curve, roc_auc_score
from sklearn.metrics import PrecisionRecallDisplay

# Montar Google Drive si estás usando Google Colab
from google.colab import drive
drive.mount('/content/gdrive')

# Importar el dataset
df = pd.read_csv('/content/gdrive/MyDrive/Colab Notebooks/creditcard.csv')

# Limpieza de datos y análisis exploratorio

# Observar las primeras filas del dataframe
print(df.head())

# Verificar la información general del dataset
print(df.info())

# Verificar la existencia de valores nulos en el dataset
print(df.isnull().sum())

# Estadísticas descriptivas del dataset
print(df.describe())

# Separar características (X) y etiquetas (y)
X = df.drop('Class', axis=1)
y = df['Class']

# Dividir el dataset en entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Aplicar SMOTE para balancear las clases en el conjunto de entrenamiento
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

# Normalizar los datos
scaler = StandardScaler()
X_resampled = scaler.fit_transform(X_resampled)
X_test = scaler.transform(X_test)

# Crear y entrenar el modelo de regresión logística
logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_resampled, y_resampled)

# Predecir probabilidades
y_probs_logistic = logistic_model.predict_proba(X_test)[:, 1]

# Calcular precision-recall curve
precision, recall, thresholds = precision_recall_curve(y_test, y_probs_logistic)

# Ajustar las predicciones utilizando el umbral predeterminado si hay problemas de longitud
if len(precision) != len(recall) or len(precision) != len(thresholds):
    print("Warning: Precision, recall, and thresholds have different lengths. Using default threshold.")
    optimal_threshold = 0.5  # Usar un umbral predeterminado
else:
    optimal_idx = np.argmax(np.abs(precision - recall))
    optimal_threshold = thresholds[optimal_idx]

# Ajustar las predicciones utilizando el umbral óptimo
y_pred_logistic_thresh = (y_probs_logistic >= optimal_threshold).astype(int)

# Evaluar el modelo
print("Evaluación del Modelo de Regresión Logística Mejorado")
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_logistic_thresh))
print("\nClassification Report:")
print(classification_report(y_test, y_pred_logistic_thresh))
print(f"Accuracy: {accuracy_score(y_test, y_pred_logistic_thresh):.4f}")
print(f"AUC-ROC Score: {roc_auc_score(y_test, y_probs_logistic):.4f}")

# Visualizar la curva precision-recall
disp = PrecisionRecallDisplay(precision=precision, recall=recall)
disp.plot()
plt.title('Precision-Recall Curve')
plt.show()

````
