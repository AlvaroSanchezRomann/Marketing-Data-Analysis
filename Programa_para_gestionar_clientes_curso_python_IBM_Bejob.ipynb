```python
class Cliente:
    def __init__(self, nombre, email, telefono):
        self.nombre = nombre
        self.email = email
        self.telefono = telefono

    def __str__(self):
        return f"Nombre: {self.nombre}\nEmail: {self.email}\nTeléfono: {self.telefono}"


class SistemaClientes:
    def __init__(self):
        self.clientes = []

    def agregar_cliente(self, cliente):
        self.clientes.append(cliente)
        return "Cliente agregado con éxito."

    def eliminar_cliente(self, nombre):
        cliente = self.buscar_cliente(nombre)
        if cliente:
            self.clientes.remove(cliente)
            return "Cliente eliminado con éxito."
        else:
            return "No se encontró el cliente."

    def buscar_cliente(self, nombre):
        for cliente in self.clientes:
            if cliente.nombre == nombre:
                return cliente
        return None

    def listar_clientes(self):
        if not self.clientes:
            return "No hay clientes registrados."
        else:
            return "\n".join(str(cliente) + "\n" + "-"*20 for cliente in self.clientes)


# Función para mostrar el menú y manejar las opciones en Google Colab
def manejar_menu(opcion, sistema, **kwargs):
    if opcion == "1":
        nombre = kwargs.get("nombre")
        email = kwargs.get("email")
        telefono = kwargs.get("telefono")
        cliente = Cliente(nombre, email, telefono)
        return sistema.agregar_cliente(cliente)

    elif opcion == "2":
        nombre = kwargs.get("nombre")
        return sistema.eliminar_cliente(nombre)

    elif opcion == "3":
        nombre = kwargs.get("nombre")
        cliente = sistema.buscar_cliente(nombre)
        if cliente:
            return str(cliente)
        else:
            return "No se encontró el cliente."

    elif opcion == "4":
        return sistema.listar_clientes()

    elif opcion == "5":
        return "Saliendo del sistema. ¡Hasta luego!"

    else:
        return "Opción no válida. Por favor, seleccione una opción entre 1 y 5."


# Crear una instancia del sistema de clientes
sistema = SistemaClientes()

# Prueba 1: Agregar un cliente
print(manejar_menu("1", sistema, nombre="Juan Pérez", email="juan@example.com", telefono="123456789"))

# Verificar cliente agregado
print(manejar_menu("4", sistema))

# Prueba 2: Buscar un cliente
print(manejar_menu("3", sistema, nombre="Juan Pérez"))

# Prueba 3: Eliminar un cliente
print(manejar_menu("2", sistema, nombre="Juan Pérez"))

# Verificar eliminación
print(manejar_menu("4", sistema))

# Prueba 4: Intentar buscar cliente eliminado
print(manejar_menu("3", sistema, nombre="Juan Pérez"))

# Prueba 5: Intentar eliminar un cliente que no existe
print(manejar_menu("2", sistema, nombre="Ana Martínez"))

# Prueba 6: Listar clientes cuando no hay ninguno
print(manejar_menu("4", sistema))

# Pruebas adicionales
print("\n--- Pruebas adicionales ---")

# Prueba 1: Agregar múltiples clientes
print(manejar_menu("1", sistema, nombre="Juan Pérez", email="juan@example.com", telefono="123456789"))
print(manejar_menu("1", sistema, nombre="Ana Martínez", email="ana@example.com", telefono="987654321"))
print(manejar_menu("1", sistema, nombre="Luis García", email="luis@example.com", telefono="555555555"))

# Verificar clientes agregados
print(manejar_menu("4", sistema))

# Prueba 2: Buscar un cliente existente
print(manejar_menu("3", sistema, nombre="Ana Martínez"))

# Prueba 3: Intentar eliminar un cliente que no existe
print(manejar_menu("2", sistema, nombre="Carlos López"))

# Intentar eliminar clientes que existen
print(manejar_menu("2", sistema, nombre="Luis García"))
print(manejar_menu("2", sistema, nombre="Ana Martínez"))

# Verificar eliminación de todos los clientes
print(manejar_menu("4", sistema))

# Prueba 4: Intentar eliminar y buscar clientes cuando no hay ninguno
print(manejar_menu("2", sistema, nombre="No Existe"))
print(manejar_menu("3", sistema, nombre="No Existe"))
print(manejar_menu("4", sistema))

```
